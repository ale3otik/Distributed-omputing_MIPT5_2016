```python
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
#nsections = 1E8
nthr = np.arange(1,13)

t = [1.321128, 0.670762, 0.458050, 0.346426, 0.283829, 0.241168, 0.209002, 0.185471, 0.166970, 0.153814, 0.147310, 0.135790]

t1_crit = np.array([3.766820, 1.617997, 1.003034, 0.755241, 0.652278, 0.608260, 0.514275, 0.474068, 0.423287, 0.376195, 0.344293, 0.320153])
t2_red = np.array([3.017742, 1.490175, 0.991173, 0.764051, 0.627146, 0.533261, 0.447214, 0.400170, 0.354539, 0.339743, 0.305035, 0.304906])
```

```python
plt.figure(figsize=(15,6))
plt.grid(True)
plt.plot(nthr,t1_crit,'-o',color='red',label='critical')
plt.plot(nthr,t2_red,'-o',color='green',label='reduce')
plt.legend()
plt.show()
```

```python
boost_crit = [t1_crit[0]/x for x in t1_crit]
boost_red = [t2_red[0]/x for x in t1_crit]

#nsections = 1E5
t_st = np.array([0.003524, 0.299317, 0.385941, 0.397328, 0.391091, 0.403398, 0.403074, 0.399459, 0.404006, 0.420355, 0.403035, 0.414957])
boost_st = t_st[0] / t_st;


plt.figure(figsize=(15,6))
plt.grid(True)
plt.plot(nthr,boost_st,'-o',color='blue',label='critical_stupid')
plt.plot(nthr,boost_crit,'-o',color='red',label='critical')
plt.plot(nthr,boost_red,'-o',color='green',label='reduce')
plt.legend(loc="best")
plt.savefig("plot.png")
plt.xlabel('nthreads')
plt.ylabel('acceleration')
plt.show()
```
