```python
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
nthr = np.arange(1,13)
t1_crit = np.array([0.163626, 0.081555, 0.077279, 0.074166, 0.077821, 0.073363, 0.075120, 0.074813, 0.073945, 0.075871, 0.076295, 0.076789])
t2_red = np.array([0.167205, 0.087695, 0.083258, 0.076082, 0.075536, 0.074341, 0.077735, 0.074881, 0.075600, 0.081362, 0.078164, 0.075785])
```

```python
plt.figure(figsize=(15,6))
plt.grid(True)
plt.plot(nthr,t1_crit,'-o',color='red',label='critical')
plt.plot(nthr,t2_red,'-o',color='green',label='reduce')
plt.legend()
plt.show()
```

```python
boost_crit = [t1_crit[0]/x for x in t1_crit]
boost_red = [t2_red[0]/x for x in t1_crit]

plt.figure(figsize=(15,6))
plt.grid(True)
plt.plot(nthr,boost_crit,'-o',color='red',label='critical')
plt.plot(nthr,boost_red,'-o',color='green',label='reduce')
plt.legend()
plt.show()
```
